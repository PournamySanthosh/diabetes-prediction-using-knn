```python
#recursion
#a function is called inside the same function
def fact(n):
  if n==1 or n==0:
    return 1
  else:
    return n*fact(n-1)
fact(5)
```

```python
#wap to calculate sum of natural nos using recursion
def sum(n):
  if n<=1:
    return n
  else:
    return n+sum(n-1)
sum(6)
```

```python
#find the power of two numbers using recusrion
def power(a,b):
  if b==0:
    return 1
  else:
    return a*power(a,b-1)
print(power(6,6))
```

```python
class Student():
  def __init__(self,name,rollno):
    self.name=name
    self.rollno=rollno

  def display_data(self):
    print("name = ",self.name)
    print("roll number = ",self.rollno)

class Test(Student):
  def __init__(self, name, rollno,mark):
    super().__init__(self,name, rollno)
    self.mark=mark

  def display_mark(self):
    print("Mark=",self.mark)

class Sports(Student):
  def __init__(self, name, rollno, smark):
    Student.__init__(self, name, rollno)
    self.smark = smark

  def display_smark(self):
    print("SMark=",self.smark)

class Grade(Test,Sports):
  def __init__(self, name, rollno, mark,smark):
    Test.__init__(self,name, rollno, mark)
    Sports.__init__(self, name, rollno, smark)
    self.mark=mark
    self.smark=smark


  def calgrade(self):
    p = self.mark / 500 * 100
    if p>80:
      print("Grade A")
    elif p>50:
      print("Grade B")
    elif p>30:
      print("Grade C")
    else:
      print("Fail")

g1 = Grade("Pournamy", 45, 70, 80)
g1.display_data()
g1.display_mark()
g1.display_smark()
g1.calgrade()
```

```python
class Point():
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def __add__(self,other):
    x=self.x+other.x
    y=self.y+other.y
    return Point(x,y)
  def __str__(self):
    return "({0},{1})".format(self.x,self.y)
p1=Point(1,2)
p2=Point(3,4)
print(p1+p2)
```

```python
#Consider a list (list = []). You can perform the following commands:

#insert i e: Insert e integer  at i position .
#print: Print the list.
#remove e: Delete the first occurrence of integer .
#append e: Insert integer  at the end of the list.
#sort: Sort the list.
#pop: Pop the last element from the list.
#reverse: Reverse the list.
list=[]
n=int(input())

for i in range(n):
  comm=input().split()
  if comm[0]=="insert":
    i=int(comm[1])
    e=int(comm[2])
    list.insert(i,e)
  elif comm[0]=="print":
    print(list)
  elif comm[0]=="remove":
    i=int(comm[1])
    list.remove(i)
  elif comm[0]=="append":
    i=int(comm[1])
    list.append(i)
  elif comm[0]=="sort":
    list.sort()
  elif comm[0]=="pop":
    list.pop()
  elif comm[0]=="reverse":
    list.reverse()


```

```python
import pandas as pd
df=pd.read_csv("/content/diabetes.csv")
```

```python
#eda
df.head()
df.info()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,0:8]
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_train
```

```python
x_test=scaler.transform(x_test)
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
y_predict=knn.predict(x_test)
```

```python
y_test.values
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_predict)*100
ac
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,y_predict)
cm
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
cm_display=ConfusionMatrixDisplay.from_predictions(y_test,y_predict)
cm
```

```python
from sklearn.metrics import classification_report
print(cm)
cr=classification_report(y_test,y_predict)
print(cr)
```
